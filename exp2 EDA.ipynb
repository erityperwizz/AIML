```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r'/content/cancer patient data sets.csv')
df.head()
```

```python
# checking missing values if present
print(df.isnull().sum()) # No missing values hence present
```

```python
df1 = df
df1['Duplicated'] = df.duplicated(keep=False) # keep=False marks the duplicated row with a True
df_final = df1[~df1['Duplicated']] # selects only rows which are not duplicated.
print(df1.size)
print(df_final.size) # output says no rows are duplicated as size remains the same
```

```python
# Generate summary statistics for relevant numerical variables (e.g., mean, median, standard deviation, etc.)
df.describe()
```

```python
# Visulizing using histograms
df.hist(column='Age',bins=10,by='Gender')
plt.xlabel('Sex')
plt.ylabel('Number of Patients')
plt.show() # 1=Female, 2=Male
```

```python
# Visualize the correlation between variables using a heatmap
plt.figure(figsize=(18, 18))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Genetic Risk and CHRONIC LUNG DISEASE
# Visualize the relationship between age and maximum heart rate achieved
plt.figure(figsize=(10, 6))
sns.barplot(x='Genetic Risk', y='chronic Lung Disease', hue='Gender',palette = "Set1",data=df)
plt.title('LUNG DISEASE')
plt.xlabel('Genetic Risk')
plt.ylabel('CHRONIC LUNG DISEASE')
plt.show()
```
